```python
import pandas as pd
from sqlalchemy import create_engine

# Database connection parameters
DB_USER = "postgres"
DB_PASSWORD = "123456"
DB_HOST = "localhost"
DB_PORT = "5432"
DB_NAME = "ecopackai_db"

# Create Database Connection Engine
try:
    connection_str = f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
    engine = create_engine(connection_str)
    print("Successfully connected to the database!")
except Exception as e:
    print(f"Error connecting to database: {e}")
```

```python
# Query 'materials' table
try:
    print("\n--- Materials Table Data ---")
    query_materials = "SELECT * FROM materials LIMIT 10;"
    df_materials = pd.read_sql(query_materials, engine)
    display(df_materials)
except Exception as e:
    print(f"Error querying materials table: {e}")
```

```python
# Query 'product_material' table
try:
    print("\n--- Product/Material Table Data ---")
    query_pm = "SELECT * FROM product_material LIMIT 10;"
    df_pm = pd.read_sql(query_pm, engine)
    display(df_pm)
except Exception as e:
    print(f"Error querying product_material table: {e}")
```
