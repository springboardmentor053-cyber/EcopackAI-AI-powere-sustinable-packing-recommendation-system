**Handling Missing Values**

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object", "bool"]).columns

# Fill numeric with mean
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

# Fill categorical with mode
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

# **Feature Engineering**

```python
# 1. CO2 Impact Index (lower CO2 is better)
df["co2_impact_index"] = 1 / (df["co2_emission_kg"] + 1)

# 2. Cost Efficiency Index (you already have cost_efficiency, but we refine it)
df["cost_efficiency_index"] = df["strength_score"] / (df["cost_per_unit_usd"] + 1)

# 3. Material Suitability Score (main ranking feature)
df["material_suitability_score"] = (
    0.35 * df["strength_score"] +
    0.25 * df["biodegradability_score"] +
    0.20 * (df["recyclability_percent"] / 100) +
    0.20 * df["reuse_efficiency"]
)
df.head()
```

```python
ordinal_map = {"Low": 1, "Medium": 2, "High": 3}

# Ordinal encoding
df["moisture_resistance"] = df["moisture_resistance"].map(ordinal_map)
df["heat_resistance"] = df["heat_resistance"].map(ordinal_map)
```

```python
df["compostable"] = df["compostable"].map({"Yes": 1, "No": 0})
```

```python
categorical_cols = ["material_name", "industry_standard"]

df_encoded = pd.get_dummies(
    df,
    columns=categorical_cols,
    drop_first=True
)
```

```python
non_numeric = df_encoded.select_dtypes(include=["object"]).columns
print("Non-numeric columns:", non_numeric)
```

```python
X = df_encoded.drop(["cost_per_unit_usd", "co2_emission_kg"], axis=1)

y_cost = df_encoded["cost_per_unit_usd"]
y_co2  = df_encoded["co2_emission_kg"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_cost_train, y_cost_test = train_test_split(
    X, y_cost, test_size=0.2, random_state=42
)

_, _, y_co2_train, y_co2_test = train_test_split(
    X, y_co2, test_size=0.2, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

rf_model.fit(X_train_scaled, y_cost_train)
```

```python
!pip install xgboost
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(
    n_estimators=200,
    random_state=42
)

xgb_model.fit(X_train_scaled, y_co2_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_cost_pred = rf_model.predict(X_test_scaled)

print("Cost Prediction Model")
print("MAE :", mean_absolute_error(y_cost_test, y_cost_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_cost_test, y_cost_pred)))
print("R2  :", r2_score(y_cost_test, y_cost_pred))
```

```python
y_co2_pred = xgb_model.predict(X_test_scaled)

print("CO2 Prediction Model")
print("MAE :", mean_absolute_error(y_co2_test, y_co2_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_co2_test, y_co2_pred)))
print("R2  :", r2_score(y_co2_test, y_co2_pred))
```

```python
results = X_test.copy()

results["predicted_cost"] = y_cost_pred
results["predicted_co2"] = y_co2_pred

# Final eco-ranking score
results["final_score"] = (
    0.5 * (1 / (results["predicted_cost"] + 1)) +
    0.5 * (1 / (results["predicted_co2"] + 1))
)

ranked_materials = results.sort_values(by="final_score", ascending=False)

ranked_materials.head(10)
```

```python
ranked_materials.to_sql(
    "material_recommendations",
    engine,
    if_exists="replace",
    index=False
)
```
