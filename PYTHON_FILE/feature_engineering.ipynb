```python
import pandas as pd
```

```python
material_df = pd.read_csv(r"D:\EcopackAI Project\DATA\material.csv")
```

```python
material_df["co2_impact_index"] = (
    material_df["co2_emission_score"] * 0.5 +
    (100 - material_df["recyclability_percent"]) * 0.3 +
    (10 - material_df["biodegradability_score"]) * 0.2
)

material_df[[
    "co2_emission_score",
    "recyclability_percent",
    "biodegradability_score",
    "co2_impact_index"
]].head()
```

```python
material_df["performance_index"] = (
    material_df["strength_score"] * 0.6 +
    material_df["weight_capacity_score"] * 0.4
)

material_df[[
    "strength_score",
    "weight_capacity_score",
    "performance_index"
]].head()
```

```python
material_df["material_suitability_score"] = (
    material_df["performance_index"] * 0.35 +
    material_df["biodegradability_score"] * 0.25 +
    (material_df["recyclability_percent"] / 100) * 0.2 -
    material_df["co2_emission_score"] * 0.2
)

material_df[["material_suitability_score"]].head()
```
