```python
import pandas as pd
import numpy as np
```

```python
materials = pd.read_csv("E:\Computer Science\Intern\EcoPackAI\EcopackAI-AI-powere-sustinable-packing-recommendation-system\Dataset\eco_packaging_materials.csv")
products = pd.read_csv("E:\Computer Science\Intern\EcoPackAI\EcopackAI-AI-powere-sustinable-packing-recommendation-system\Dataset\product_categories.csv")
```

```python
num_cols = [
    'strength_score', 'weight_capacity_kg',
    'biodegradability_score', 'recyclability_percent',
    'co2_emission_kg', 'cost_per_kg'
]

materials[num_cols] = materials[num_cols].fillna(materials[num_cols].mean())
```

```python
cat_cols = ['material_type', 'water_resistance', 'industry_usage']
materials[cat_cols] = materials[cat_cols].fillna(materials[cat_cols].mode().iloc[0])
```

```python
materials['food_safe'] = materials['food_safe'].fillna(False)
```

```python
materials.drop_duplicates(inplace=True)
products.drop_duplicates(inplace=True)
```

```python
materials.isnull().sum()
materials.duplicated().sum()
```

```python
materials.to_csv("materials_cleaned.csv", index=False)
products.to_csv("products_cleaned.csv", index=False)
```
