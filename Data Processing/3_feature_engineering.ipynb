```python
import pandas as pd
import numpy as np
materials = pd.read_csv("materials_cleaned.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
materials['material_type_enc'] = le.fit_transform(materials['material_type'])
materials['water_resistance_enc'] = le.fit_transform(materials['water_resistance'])
```

```python
from sklearn.preprocessing import MinMaxScaler

num_cols = [
    'strength_score', 'weight_capacity_kg',
    'biodegradability_score', 'recyclability_percent',
    'co2_emission_kg', 'cost_per_kg'
]

scaler = MinMaxScaler()
materials[num_cols] = scaler.fit_transform(materials[num_cols])
```

```python
materials['co2_impact_index'] = 1 - materials['co2_emission_kg']
```

```python
materials['cost_efficiency_index'] = 1 - materials['cost_per_kg']
```

```python
materials['cost_efficiency_index'] = 1 - materials['cost_per_kg']
```

```python
materials['material_suitability_score'] = (
    0.30 * materials['biodegradability_score'] +
    0.25 * materials['recyclability_percent'] +
    0.20 * materials['strength_score'] +
    0.15 * materials['co2_impact_index'] +
    0.10 * materials['cost_efficiency_index']
)
```

```python
materials.to_csv("materials_ml_ready.csv", index=False)
```
