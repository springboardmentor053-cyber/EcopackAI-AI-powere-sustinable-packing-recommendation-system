```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
materials = pd.read_csv("D:\Project ecopackai\data\material_dataset.csv")
products = pd.read_csv("D:\Project ecopackai\data\product_dataset.csv")

print("Datasets loaded successfully")
```

```python
print(materials.isnull().sum())
print(products.isnull().sum())
```

```python
materials.fillna(materials.mean(numeric_only=True), inplace=True)
products.fillna(products.mean(numeric_only=True), inplace=True)
```

```python
materials.drop_duplicates(inplace=True)
products.drop_duplicates(inplace=True)
```

```python
if "material_type" in materials.columns:
    materials = pd.get_dummies(materials, columns=["material_type"], drop_first=True)

if "product_category" in products.columns:
    products = pd.get_dummies(products, columns=["product_category"], drop_first=True)
```

```python
score_cols = [
    "strength",
    "weight_capacity",
    "biodegradability_score",
    "co2_emission_score"
]


for col in score_cols:
    materials[col] = pd.to_numeric(materials[col], errors="coerce")
```

```python
materials[score_cols] = materials[score_cols].fillna(
    materials[score_cols].mean()
)
```

```python
materials["durability_score"] = pd.to_numeric(
    materials.get("durability_score", 5),
    errors="coerce"
)
materials["durability_score"].fillna(5, inplace=True)
```

```python
scaler = MinMaxScaler(feature_range=(1, 10))
materials[score_cols] = scaler.fit_transform(materials[score_cols])
```

```python
materials["co2_emission_score"] = 11 - materials["co2_emission_score"]
```

```python
cost_scaler = MinMaxScaler(feature_range=(1, 10))
materials["cost_score"] = cost_scaler.fit_transform(
    materials[["cost_per_unit"]]
)
materials["cost_score"] = 11 - materials["cost_score"]
```

```python
materials["co2_impact_index"] = (
    materials["biodegradability_score"] +
    materials["co2_emission_score"] +
    (materials["recyclability_percent"] / 10)
) / 3
```

```python
materials["strength"] = 5
```

```python
materials["strength"].isnull().sum()
```

```python
materials["material_suitability_score"] = materials[[
    "strength",
    "weight_capacity",
    "durability_score",
    "co2_impact_index",
    "cost_score"
]].mean(axis=1)
```

```python

materials.to_csv("materials_module2_final.csv", index=False)


products.to_csv("products_module2_final.csv", index=False)

print("✅ Datasets saved successfully")
```
