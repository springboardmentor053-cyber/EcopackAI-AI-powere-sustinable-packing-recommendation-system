```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
materials = pd.read_csv("D:\Project ecopackai\data\materials_module2_final.csv")
print(materials.head())
```

```python
feature_cols = [
    "strength",
    "weight_capacity",
    "durability_score",
    "biodegradability_score",
    "recyclability_percent"
]

X = materials[feature_cols]
```

```python
y_cost = materials["cost_per_unit"]
```

```python
y_co2 = materials["co2_emission_score"]
```

```python
X_train, X_test, y_cost_train, y_cost_test = train_test_split(
    X, y_cost,
    test_size=0.2,
    random_state=42
)

_, _, y_co2_train, y_co2_test = train_test_split(
    X, y_co2,
    test_size=0.2,
    random_state=42
)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import numpy as np

np.save("X_train.npy", X_train_scaled)
np.save("X_test.npy", X_test_scaled)

np.save("y_cost_train.npy", y_cost_train)
np.save("y_cost_test.npy", y_cost_test)

np.save("y_co2_train.npy", y_co2_train)
np.save("y_co2_test.npy", y_co2_test)

print(" Module 3 ML datasets saved successfully")
```
